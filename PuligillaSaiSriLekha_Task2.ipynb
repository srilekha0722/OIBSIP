```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("unemployment.csv")
print(df.head())
print(df.columns)
print(df.isnull().sum())
df.columns = df.columns.str.strip()
print(df.info())
plt.figure(figsize=(10,5))
sns.histplot(df['Estimated Unemployment Rate (%)'], bins=20, kde=True)
plt.title('Distribution of Unemployment Rate')
plt.show()
plt.figure(figsize=(12,8))
state_unemployment = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values()
sns.barplot(x=state_unemployment.values, y=state_unemployment.index)
plt.title('Average Unemployment Rate by Region')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Region')
plt.show()
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True,
            cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
